### Coding Q1 : Find Peak Element

```python
# Input: nums = [1, 2, 3, 1]
# Output: 2 (index 2, value 3 is a peak)
# Input: nums = [1, 2, 1, 3, 5, 6, 4]
# Output: 5 (index 5, value 6 is a peak; index 1 also valid)
```

```python
def find_peak_element(nums):
    n = len(nums)
    for i in range(n):
        left = nums[i-1] if i > 0 else float('-inf')
        right = nums[i+1] if i < n-1 else float('-inf')

        if nums[i] > left and nums[i] > right:
            return i        
```

```python
nums = [1, 2, 3, 1]
print(find_peak_element(nums))
```

### Coding Q2 : Group Anagrams

```python
input = ["eat","tea","tan","ate","nat","bat"]
output = [["bat"],["nat","tan"],["eat","ate","tea"]]
```

```python
def group_anagram(array):
    temp_dict = {}

    for word in array:
        key = "".join(sorted(word))
        if key not in temp_dict:
            temp_dict[key]=[word]
        else:
            temp_dict[key].append(word)
    return list(temp_dict.values())

print(group_anagram([""]))
```

### Coding Q3 :Kth Largest Element in an Array

```python
# Input: nums = [3, 2, 1, 5, 6, 4], k = 2
# Output: 5
# Input: nums = [3, 2, 3, 1, 2, 4, 5, 5, 6], k = 4
# Output: 4
```

```python
def kth_largest_element(nums,k):
    n = len(nums)
    for i in range(n):
        for j in range(i+1,n):
            if nums[i] <nums[j]:
                nums[i],nums[j] = nums[j],nums[i]

    return nums[k-1]
```

```python
nums = [3, 2, 3, 1, 2, 4, 5, 5, 6]
k = 4
print(kth_largest_element(nums,k))
```
